```python
def longest_word(words):
    longest = words[0]
    for i in words:
        if len(i) > len(longest):
            longest = i
    return longest
```

```python
list2 = [ 1,2,2,3,3,4,5,5]
```

```python
def remove_dup(list2):
    new_list = []
    for i in list2:
        if i not in new_list:
            new_list.append(i)
    return new_list
```

```python
print(remove_dup(list2))
```

```python
nums = [ 12,44,66,77]
```

```python
def lar_sec(nums):
    largest = nums [0]
    second = nums [0]
    for i in nums:
        if i > largest:
            second = largest
            largest = i
        elif i > second and i != largest:
            second = i
    return second
```

```python
print(lar_sec(nums))
```
