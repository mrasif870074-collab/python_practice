```python
# nesting
pin = 1234
mobile = 1111
balance = 1000
```

```python
entered_pin = int(input("enter your pin: "))
if entered_pin == pin:
    print("correct pin go ahead")
    entered_mobile = int(input("enter your mobile: "))
    if entered_mobile == mobile:
        print("correct mobile number go ahead")
        amount = int(input("enter your amount: "))
        if amount <= balance:
            print("transsaction successful")
            print("remaining balance",balance-amount)
            balance = balance-amount
        else:
            print("you dont have balance in your account")
    else:
        print("mobile does not match")

else:
    print("incorrect pin 75468")
```
