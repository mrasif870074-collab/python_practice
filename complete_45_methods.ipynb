# All Python String Methods

This notebook contains all public Python string methods with multiple examples and comments for different arguments.

# 1. string.capitalize()

Converts the first character to uppercase and the rest to lowercase.

```python
my_string = "python"
print(my_string.capitalize()) # First letter becomes capital

my_string = "pYTHON PROGRAMMING"
print(my_string.capitalize()) # First letter capital, remaining letters lowercase


my_string = "123 python"
print(my_string.capitalize()) # Number is unchanged, first alphabet stays same place
```

# 2. string.casefold()

Converts the string to lowercase more strongly than lower(), useful for case-insensitive comparison.

```python
my_string = "PYTHON"
print(my_string.casefold()) # Converts uppercase letters to lowercase

my_string = "Python Programming"
print(my_string.casefold()) # Converts mixed case to lowercase

first = "PYTHON"
second = "python"
print(first.casefold() == second.casefold()) # Checks both strings without case difference
```

# 3. string.center(width[, fillchar])

Returns a centered string using spaces or the given fill character.

```python
my_string = "Python"
print(my_string.center(12)) # Center string in width 12 using spaces

my_string = "Python"
print(my_string.center(12, "-")) # Center string in width 12 using -

my_string = "Hi"
print(my_string.center(10, "*")) # Center string in width 10 using *
```

# 4. string.count(substring[, start[, end]])

Counts how many times a substring appears in the string.

```python
my_string = "banana"
print(my_string.count("a")) # Count all a characters

my_string = "banana"
print(my_string.count("a", 2)) # Count a from index 2 to end

my_string = "banana"
print(my_string.count("a", 2, 5)) # Count a from index 2 to before index 5

my_string = "Python is easy and Python is powerful"
print(my_string.count("Python")) # Count word Python
```

# 5. string.encode(encoding='utf-8', errors='strict')

Converts the string into bytes using the given encoding.

```python
my_string = "Python"
print(my_string.encode()) # Encode using default utf-8

my_string = "Python"
print(my_string.encode("utf-8")) # Encode using utf-8

my_string = "Hello123"
print(my_string.encode("ascii")) # Encode using ascii

my_string = "cafe" + chr(233)
print(my_string.encode("ascii", errors="ignore")) # Ignore characters ascii cannot encode
```

# 6. string.endswith(suffix[, start[, end]])

Checks whether the string ends with the given suffix.

```python
my_string = "welcome.py"
print(my_string.endswith(".py")) # Is string ending with .py?

my_string = "image.jpeg"
print(my_string.endswith((".jpg", ".jpeg", ".png"))) # Is string ending with any image extension?

my_string = "Python Programming"
print(my_string.endswith("Programming", 7)) # Is Programming ending from index 7 to end?

my_string = "Python Programming"
print(my_string.endswith("Program", 7, 14)) # Is Program ending between index 7 and 14?
```

# 7. string.expandtabs(tabsize=8)

Replaces tab characters with spaces.

```python
my_string = "Name\tAge"
print(my_string.expandtabs()) # Replace tab using default tab size 8

my_string = "Name\tAge"
print(my_string.expandtabs(4)) # Replace tab using tab size 4

my_string = "A\tB\tC"
print(my_string.expandtabs(2)) # Replace each tab using tab size 2
```

# 8. string.find(substring[, start[, end]])

Returns the first index of a substring. If it is not found, returns -1.

```python
my_string = "hello world"
print(my_string.find("world")) # Find world in full string

my_string = "banana"
print(my_string.find("a", 2)) # Find a from index 2

my_string = "banana"
print(my_string.find("a", 2, 5)) # Find a from index 2 to before index 5

my_string = "Python"
print(my_string.find("Java")) # Returns -1 because Java is not found
```

# 9. string.format(*args, **kwargs)

Formats values into placeholders inside the string.

```python
my_string = "My name is {}"
print(my_string.format("Sachin")) # Put value in empty placeholder

my_string = "{} is learning {}"
print(my_string.format("Sachin", "Python")) # Put values in order

my_string = "{name} is learning {course}"
print(my_string.format(name="Riya", course="Python")) # Put values by name

my_string = "Total marks: {:.2f}"
print(my_string.format(95.678)) # Show number with 2 decimal places
```

# 10. string.format_map(mapping)

Formats placeholders using values from a dictionary-like mapping.

```python
student = {"name": "Riya", "course": "Python"}
print("{name} is learning {course}".format_map(student)) # Fill placeholders from dictionary

marks = {"subject": "Math", "score": 92}
print("{subject}: {score}".format_map(marks)) # Use subject and score keys

employee = {"name": "Sachin", "city": "Pune"}
print("{name} lives in {city}".format_map(employee)) # Use name and city keys
```

# 28. string.maketrans(x[, y[, z]])

Creates a translation table for translate(). It can replace characters, delete characters, or use a dictionary mapping.

```python
table = str.maketrans("is", "at")
text = "This is a cat"
result = text.translate(table)
print(result) # i is replaced with a and s is replaced with t

translation_table = str.maketrans("abc", "123", "x")
text = "abcdex"
result = text.translate(translation_table)
print(result) # a, b, c are replaced with 1, 2, 3 and x is deleted

translation_table = str.maketrans("", "", "aei")
text = "hello, amazing world!"
result = text.translate(translation_table)
print(result) # Delete characters a, e, and i

translation_table = str.maketrans({"a": "@", "b": "8", "c": "("})
text = "abcde"
result = text.translate(translation_table)
print(result) # a becomes @, b becomes 8, and c becomes (
```
