# Dictionary 

# what is dictionary?
# A dictionary in python is a collection of key value pairs.

why dictionary?
Dictionary are highly efficient for lookup, modification, and deletion operations,
making them ideal for scenarios where data need to be accessed using meaningful
identifiers rather than numeric indices.

# Dictionary characteristics

1. ordered or unordered: maintains the insertion order of elements.

    . before python 3.7, ditionaries did not maintain any order of items.

    . from python 3.7+, dictioaries preserve the insertion order of keys.

    . when you itrate over a dictionary (e.g., using a for loop or methods like.
      keys(),.values(),or .items(), the elements are returned in the same order as
      they were inserted into the dictionary.

2. unindexed: elements are accessed via keys, not indices.

3. mutable: yes, in python, dictionaries are mutable. this means you can modify a dictionary after it has been created, such as adding, removing, or changing elements.

4. must be immutable and unique:

   . dictionary keys must be immutable, meaning they can be of types likes:

   .strings

   .Numbers (integers or floats)

   .Tuples(if all elements of the tuple are immutable)

   List, sets, or dictionaries cannot be used as keys.

5. Efficient Lookup:
   dictionaries offer O(1) averange time complexity for lookups, insertions, and deletions because they use a hash table under the hood.
