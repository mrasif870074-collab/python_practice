# Tuple characteristics

1. ordered: mainatains the order of elements.

    Elements in a tuple have a defined order, which means you can access each element by its index,
     Starting from 0 for the first element.

                                                       List indexing
                                          #    Index    Index   Index     Index      Index
                                          #      0        1       2         3           4
                                          # [["HELLO"], [122], ["Mysql"], ["Python"], [2.2]]
   
   2. Indexed: Elements in a tuple can be accessed using their index,
              starting from 0 for the first element.

   3. Immutable: once a tuple is created, its elements cannot be changed, modified, or deleted.
  
    This immutability makes tuples useful for storing data that should remain constant throughout the program.

   4. Allows Duplicate Elements:
   tuple can contain duplicate values, just like list.

   my_tuple = (1, 2, 2, 3)

   
5. Heterogeneous:
. tuple can store elements of different data types(e.g, integers,strings,lists,etc.).

             my_tuple = (1,"python", (3,4,5))

what is hashable? 

Hashable refers to an object that has a fixed, unique hash value during its lifetime.

Hashing is the process of converting data into a fixed - size integer (a hash code) that
uniquely represents the object.


Hashable object must satisfy these two criteria 

1. Immutabaility:
2. Equality Consistency:

   if two object are considered equal (obh1 == obj2)

   hash values must also be equal (hash(obj1) == hash(obj2))


   example of hashable object

Numbers: int,float,complex

Tuples: (1,2,3) (all elements are hashable)

strings


Frozensets: Immutable sets.


    Example of Non- Hashable objects
    
List: [1,2,3]

DIctionaries: {"key": "values"}

sets: {1,2,3}
