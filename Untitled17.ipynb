```python
# positive indexing                      H E L L O
#                                        0 1 2 3 4


string = "hello"


print(string[0])
print(string[1])
print(string[2])
print(string[3])
print(string[4])
```

```python
string = "hello"                             # H  E L L O
                                             # -5 -4-3-2-1

print(string[-5])
print(string[-4])
print(string[-3])
print(string[-2])
print(string[-1])
```

```python
# python slicing positive indexing

# print(word[start:stop:step])

# start = 0                                   H E L L O
# stop = len(word)                            0 1 2 3 4
# step = 1  

name = "hello"

print(name[2:4:1])
print(name[0:2:1])
print(name[0:3:2])
print(name[1::4])
```

```python
# negative indexing                          H  E L L O
                                           # -5 -4-3-2-1
name = "hello"

print(name[-5:-1:1])
print(name[-4:-2:1])
print(name[-1::])
print(name[-5::1])
```

```python
name = "madhav"

print(name[0:3:1])
```

```python
# string formatting 
# string formatting refers to the precess of creating string with placeholders that can be replaced by values dynamically

name = "asif"
age = 20
city = "delhi"

print ("hi name your city is city age is age")
print(f"hi {name} your city is {city} age is {age}")
print("hi {} your city is {} age is {}".format(name,age,city))
print("hi %s your city is %s age is %s"%(name,age,city))
```
