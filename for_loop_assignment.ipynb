```python
# Q no:5   Write a Python program to count the number of even and odd numbers from a
#                     series of numbers using a for loop.

for i in range(1,21):
    if i %2 ==0:
        print(i,"Even")
    else:
        print(i,"Odd")
```

```python
# Q no 6:  Write a Python program to find the factorial of a number using a for loop.

num = int(input("enter your number:"))
fact = 1
for i in range(1,num+1):
    fact = fact * i

print("factorial =",fact)
```

```python
# Q no 7: Write a Python program to print the Fibonacci sequence up to a specified number
#                                     using a for loop.

num = int(input("enter the limit:"))
a = 0
b = 1

while a <= num:
    print(a)
    c = a + b
    a = b
    b = c
```

```python
# Q no 8: Write a Python program to check if a given number is prime or not using a for loop.

num = int(input("enter your number:"))

if num <= 1:
    print("Not Prime")
else:
    for i in range(2, num):
        if num % i ==0:
            print("Not Prime")
            break
        else:
            print("Prime")
```

```python
# Q no 9: Write a Python program to find the largest element in a list using a for loop.

numbers = [10,25,8,45,30]

largest = numbers[0]

for i in numbers:
    if i > largest:
        largest = i

print("Largest number is:",largest)
```

```python
# Q no 10: Write a Python program to reverse a given string using a for loop.

text = input("enter a string:")

reverse = ""

for i in text:
    reverse = i + reverse

print("Reverse string:",reverse)
```

```python
# Q no 11: Write a Python program to find the common elements between two lists using a
#                               for loop.
#                             List1 = [1,2,3]
#                             List2 = [4,5,1]

list1 = [1,2,3]
list2 = [4,5,1]

for i in list1:
    for j in list2:
        if i == j:
            print("Common element:",i)
```
