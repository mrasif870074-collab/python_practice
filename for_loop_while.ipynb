```python
import tkinter as tk                                                # Table generator

def generate_table():
    number = int(entry.get()) # int(input("enter your number:"))
    for i in range(1,11):
        result_text.insert(tk.END,f"{number} x {i} = {number * i}\n")
        print(f"{number} x {i} = {number *i}")






root = tk.Tk()
root.title("Table Generator")
root.geometry("600x600")
input_label = tk.Label(root,text = "enter your number:")
input_label.grid(row =0,column=0,padx = 10, pady =10)

# entry box
entry = tk.Entry(root)
entry.grid(row =0,column=1,padx=10,pady=10)

#creation button 
generate_button = tk.Button(root,text="Generate Table",command=generate_table)
generate_button.grid(row=1,column=1)
# text box creation
result_text = tk.Text(root,height=18,width = 30)
result_text.grid(row=2,column=1,padx=10,pady=10)




root.mainloop()
```
